```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_circles, make_moons
from sklearn.cluster import KMeans, DBSCAN
```

# Standard clustering

```python
points, labels = make_blobs(2000, n_features=2, random_state=77)
```

```python
print('x values\t:', [round(x_i,2) for x_i in points[:,0][:5]])
print('y values\t:', [round(y_i,2) for y_i in points[:,1][:5]])
print('labels\t\t:', [label for label in labels[:5]])
```

```python
plt.figure(figsize=(5,5))
plt.scatter(points[:,0], points[:,1], s=0.7)
plt.show()
```

KMeans

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(points)
```

```python
kmeans_preds = kmeans.predict(points)
np.unique(kmeans_preds, return_counts=True)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(points[:,0], points[:,1], c=kmeans_preds, s=0.7)
plt.show()
```

DBSCAN

```python
dbscan = DBSCAN(eps=1, min_samples=5)
dbscan_preds = dbscan.fit_predict(points)
```

```python
np.unique(dbscan_preds, return_counts=True)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(points[:,0], points[:,1], c=dbscan_preds, s=0.7)
plt.show()
```

# Cluster inside cluster

```python
points, labels = make_circles(2000, noise=0.1, 
                              factor=0.2, random_state=77)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(points[:,0], points[:,1], s=0.7)
plt.show()
```

How KMeans work in the case of cluster inside cluster

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(points)
```

```python
kmeans_preds = kmeans.predict(points)
np.unique(kmeans_preds)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(points[:,0], points[:,1], c=kmeans_preds, s=0.7)
plt.show()
```

How DBSCAN work in the case of cluster inside cluster

```python
dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan_preds = dbscan.fit_predict(points)
np.unique(dbscan_preds, return_counts=True)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(points[:,0], points[:,1], c=dbscan_preds, s=0.7)
plt.show()
```

# Moons

```python
points, labels = make_moons(2000, noise=0.1, random_state=77)
plt.figure(figsize=(6,4))
plt.scatter(points[:,0], points[:,1], s=0.7)
plt.show()
```

KMeans on moons data

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(points)
kmeans_preds = kmeans.predict(points)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(points[:,0], points[:,1], c=kmeans_preds, s=0.7)
plt.show()
```

DBSCAN on moons data

```python
dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan_preds = dbscan.fit_predict(points)
np.unique(dbscan_preds, return_counts=True)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(points[:,0], points[:,1], c=dbscan_preds, s=0.7)
plt.show()
```
